```python
plt.figure(figsize = (11,7))
plt.plot(boston_df['age'].sort_values().unique(), color='green')
plt.title('Age', fontsize = 10)
plt.xlabel("Row Indexes")
plt.ylabel("AGE")
plt.show()
```

```python
plt.figure(figsize=(20,9))
plt.plot(boston_df['age'])
plt.title('Line Plot of Age', fontsize = 10)
plt.xlabel("Row Indexes")
plt.ylabel("AGE")
plt.show()
```

```python
# BAR GRAPH - BIVARIATE DATA ANALYSIS(TWO VARIABLES)
plt.figure(figsize=(15,9))
plt.bar(x = boston_df['zn'] , height = boston_df['age'], width = 1.5, color = 'green')
plt.title('Age vs ZN', fontsize = 10)
plt.xlabel("ZN")
plt.ylabel("AGE")
plt.show()
```
